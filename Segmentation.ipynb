```python
uk_data = data[data.Country=='United Kingdom']
```

```python
uk_data
```

```python
uk_data.describe()
```

```python
uk_data = uk_data[(uk_data.Quantity)>0]
```

```python
uk_data.describe()
```

```python
uk_rfm = uk_data[['CustomerID','InvoiceDate','InvoiceNo','Quantity','UnitPrice']]
```

```python
uk_rfm['TotalPrice'] = uk_rfm.loc[:,'Quantity'] * uk_rfm.loc[:,'UnitPrice']
```

```python
uk_rfm
```

```python
uk_rfm['InvoiceDate'] = pd.to_datetime(uk_rfm['InvoiceDate'])
```

```python
rfm = uk_rfm.groupby('CustomerID').agg({'InvoiceDate':lambda date:(pd.Timestamp.today(tz=None) - date.max()).days,
                                       'InvoiceNo':lambda num: len(num),
                                       'TotalPrice':lambda price: price.sum()})
```

```python
rfm.columns
```

```python
rfm.columns = ['recency','frequency','monetary']
```

```python
rfm['r_quartile'] = pd.qcut(rfm['recency'], 4, ['1','2','3','4'])
```

```python
rfm['r_quartile'] = pd.qcut(rfm['recency'], 4, ['1','2','3','4'])
rfm['f_quartile'] = pd.qcut(rfm['frequency'], 4, ['4','3','2','1'])
rfm['m_quartile'] = pd.qcut(rfm['monetary'], 4, ['4','3','2','1'])
```

```python
rfm.head()
```

```python
rfm['RFM_Score'] = rfm['r_quartile'].astype(str) + rfm['f_quartile'].astype(str) + rfm['m_quartile'].astype(str)
```
